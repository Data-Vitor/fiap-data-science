Gráfico de linhas - Ideal para séries temporais.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/datasets/anos_livros.csv')
plt.plot(df['ano'], df['livros'])
plt.title('livros lidos por ano')
plt.xlabel('ano')
plt.ylabel('livros lidos')
plt.show()
```

Gráfico de barras. - ideal para mostrar comparações entre categorias.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/datasets/dataset_leucocitos_pcr.csv')
plt.scatter(df['leucocitos'],  df['PCR'])
plt.title('Dispersão: leucitos vs. PCR')
plt.xlabel('leucitos (/mm3)')
plt.ylabel('PCR (mg/dL)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/datasets/jogadores_futebol.csv')
dfc = df[df['time'] == 'Corinthians']
v1 = dfc.groupby('ano')['cartoes_vermelhos'].sum()
plt.plot(v1)
plt.title('Cartões vermelhos conrinthinas ano')
plt.xlabel('ano')
plt.ylabel('cartoes_vermelhos')
plt.show()
```
