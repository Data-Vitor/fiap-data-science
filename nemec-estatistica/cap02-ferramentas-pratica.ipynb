Crie um gráfico de linhas que mostre, ano a ano, a quantidade total de cartões (soma de cartões amarelos e vermelhos) de todos os times. Há tendência de aumento no número de cartões ao longo do tempo? Para isso, crie uma nova variável contendo a soma dos cartões amarelos e vermelhos, e depois utilize o groupby por ano.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/datasets/jogadores_futebol.csv')

df['total_cartao'] = df['cartoes_amarelos']+ df['cartoes_vermelhos']

total_cartao = df.groupby('ano')['total_cartao'].sum()

anos = sorted(df['ano'].unique())
plt.xticks(anos)

plt.plot(total_cartao)
plt.title('Quantidade somada de cartões de todos os times')
plt.xlabel('ano')
plt.ylabel('total de cartões  por ano')
plt.show()
```
